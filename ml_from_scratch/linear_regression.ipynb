```python
import numpy as np
import matplotlib.pyplot as plt

# initialise vars for y = mx + b

np.random.seed(0)

X = np.linspace(0,10,50) # generate X points
y = 3 * X + 7 * np.random.rand(50) * 2

m = 0 
b = 0 

# initialise vars for gradient descent [m b] i+1 = [m b] i + a [d/dm d/db] i 
# and vars for sum of squared error 1/n sigma 1->n ( yi - (mxi + b))^2

learning_rate = 0.001
epochs = 1000
n = len(X)
```

```python
#Gradient Descent Alg

for i in range(epochs):

    # calculate loss 
    y_pred = m * X + b 

    loss = (y- y_pred)**2
    avg_loss = loss.sum() / n

    # calculate new m and b values 

    dm = -(2/n) * np.sum((y_pred - y) * X)
    db = -(2/n) * np.sum(y_pred - y)

    m = m + learning_rate * dm
    b = b + learning_rate * db

    #print every 100th epoch -> keep track of the run
    if i % 100 == 0:
        print(f"Epoch {i}, Loss: {avg_loss:.4f}, m: {m:.2f}, b: {b:.2f}")
    
```

```python
plt.scatter(X, y)
plt.plot(X, m*X + b, color='red')
plt.show()
```
